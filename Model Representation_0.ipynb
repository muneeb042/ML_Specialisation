First import `numpy` ( to manipulate the arrays ) and `matplotlib` to plot the different graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

 The data for the `model representation` is stored in one-dimensional NumPy arrays

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)

#Given data
x_train = np.array([1, 2 , 2.5 , 4, 6 , 3.4 , 1.7 ])
y_train = np.array([ 300, 400, 450, 600, 700, 390, 400])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

<br/> <br/>
Number of training examples = m
Numpy arrays have a `.shape parameter`. x_train.shape returns a `python tuple` with an entry for each dimension

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
#plotting the data

plt.scatter(x_train,y_train)

#setting the title
plt.title("Housing Price vs area")

#setting the labels

plt.xlabel("Area (1000sq ft)")
plt.ylabel("Price (1000K dollars)")

plt.show()
```

 <br/> <br/> <br/> <br/>The model function for linear regression (which is a function that maps from `x` to `y`) is represented as <br/> $$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

The formula above is how we can represent straight lines - different values of $w$ and $b$ give you different straight lines on the plot. <br/> <br/> 

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

Now, let's compute the value of $f_{w,b}(x^{(i)})$ for our data points. We can explicitly write this out for each data point as - 

for $x^{(0)}$, `f_wb = w * x[0] + b`

for $x^{(1)}$, `f_wb = w * x[1] + b` 

and so on... <br/>

For a large number of data points, this can get unwieldy and repetitive. So instead, we can calculate the function output in a `for` loop as shown in the `compute_model_output` function below.
> **Note**: The argument description `(ndarray (m,))` describes a Numpy n-dimensional array of shape (m,). `(scalar)` describes an argument without dimensions, just a magnitude.  
> **Note**: `np.zero(n)` will return a one-dimensional numpy array with $n$ entries   


```python
"""
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
"""
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb
    
```

Now let's call the `compute_model_output` function and plot the output..

```python
tmp_f_wb = compute_model_output(x_train,w,b,)
```

<br/>
<br/>
Plotting our `Model Prediction` below

```python
plt.plot(x_train, tmp_f_wb, c='r', label='Our prediction')
plt.scatter(x_train,y_train, label = 'Actual values')

#set the labels 
plt.xlabel("Area( in 1000Sq ft)")
plt.ylabel("Price ( in 1000k dollars)")
plt.title("Housing Price Prediction")

plt.legend()
plt.show()
```
